## Regress√£o Linear

```python
# Importando as bibliotecas

from sklearn.preprocessing import Normalizer
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
# Subindo o novo CSV a ser analisado csv

n = files.upload()
```

```python
# Importando o csv para um dataframe

df1 = pd.read_csv("dengueClimaTratados.csv")
```

```python
def converter(old):
   return int(dt.strftime(dt.strptime(old, '%Y-%m-%d'), '%m'))


for i in range(0,60,1):
    df1.Data[i] = converter(df1.Data[i])
```

```python
# Vendo as 5 primeiras entradas

df1.head()
```

```python
# Escolhendo o hyperpar√¢metro (Regress√£o Linear)

modelR = LinearRegression()
type(modelR)
```

```python
# Separando os dados entre feature matrix e target vector

X = df1.drop(columns=['CasosDengue', 'Data'], axis=1)
y = df1['CasosDengue']
```

```python
# Normalizando os dados

scaler = Normalizer().fit(X)
scaler

#scaled = scaler.transform(X)
#df1 = pd.DataFrame(scaled)
```

```python
df1.head
```

```python
# Dividindo em treino e teste

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
# Dimens√µes

print('Dimens√£o DataFrame:',df1.shape)
print('Dimens√£o X_train:', X_train.shape)
print('Dimens√£o X_test:', X_test.shape)
print('Dimens√£o y_train:', y_train.shape)
print('Dimens√£o y_test:', y_test.shape)
```

```python
# Treinando o modelo (Regress√£o Linear)
modelR.fit(X_train, y_train)

# Fazendo a previs√£o em cima dos novos dados
result_regr = modelR.predict(X_test)
```

```python
# Plotando o gr√°fico
plt.plot(X_train,y_train,color='red')
plt.plot(X_test,y_test,color='pink')
plt.plot(X_test,result_regr,color='black')
plt.show()
```

```python
# Comparando casos originais x casos previstos

plt.scatter(df1.CasosDengue, modelR.predict(X))
plt.xlabel("Casos")
plt.ylabel("Casos Previstos")
plt.show()
```

```python
# Imprimindo os coeficientes

print('Coeficiente:', modelR.coef_)
print('Intercepto:', modelR.intercept_)
```

```python
# R¬≤ (Vari√¢ncia)
print('R¬≤: ', r2_score(y_test, result_regr))

# M√©dia absoluta do erro (MAE)
print('MAE: ', mean_absolute_error(y_test, result_regr))

# M√©dia dos quadrados do erro (MSE)
print('MSE: ', mean_squared_error(y_test, result_regr))
```

```python
# Comparando casos originais x casos previstos

plt.scatter(modelR.predict(X_train), modelR.predict(X_train) - y_train, c = 'b', s = 40, alpha = 0.5)
plt.scatter(modelR.predict(X_test), modelR.predict(X_test) - y_test, c = 'g', s = 40, alpha = 0.5)
plt.hlines(y = 0, xmin = 0, xmax = 50)
plt.ylabel("Res√≠duo")
plt.title("Residual Plot - Treino(Azul), Teste(Verde)")
plt.show()
```

### Decision Tree

```python
# Treinando o modelo (Decision Tree)
modelD.fit(X_train, y_train)

# Fazendo a previs√£o no novo modelo (Decision Tree)
result_decision = modelD.predict(X_test)
```

```python
# Sa√≠das do modelo (Decision Tree)

print("Matriz de confus√£o:\n", confusion_matrix(y_test, result_decision))
print("\nScore: \n", modelD.score(X_test, y_test))
print('\nAcur√°cia: \n', accuracy_score(y_test, result_decision))
print("\nReport de Classifica√ß√£o:\n", classification_report(y_test, result_decision))
```
